### Example for Article

To start we'll need to install the Hugging Face partner package:
You can then load any Sentence Transformers model from the Hugging Face Hub.

```python
!pip install langchain-huggingface
from langchain_huggingface import HuggingFaceEmbeddings
embeddings_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-mpnet-base-v2")
```

```python
embeedings=embeddings_model.embed_query("We are writing this query ")
print(embeedings)
```

```python
embeddings = embeddings_model.embed_documents(
    [
        "Hi there!",
        "Oh, hello!",
        "What's your name?",
        "My friends call me World",
        "Hello World!"
    ]
)

print(embeddings)
```

## From loading - Embedding

```python
!pip install langchain langchain_community
from langchain.text_splitter import RecursiveCharacterTextSplitter
!pip install transformers langchain_huggingface
```

```python
!pip install pypdf
from langchain.document_loaders import PyPDFLoader
```

```python
loaders=PyPDFLoader('/content/FYP Report PhysioFlex(25july).pdf')
doc=loaders.load()
```

```python
textsplitter=RecursiveCharacterTextSplitter(
    chunk_size=300,
    chunk_overlap=50
)
split=textsplitter.split_documents(doc)
```

```python
from langchain_huggingface import HuggingFaceEmbeddings
embed_model = HuggingFaceEmbeddings(model_name='BAAI/bge-small-en-v1.5')

```

```python
embeddedquery=embed_model.embed_query("what is the scope of this project")
print(embeddedquery)
```

```python
embedded=embed_model.embed_documents([doc.page_content for doc in split])
```

```python
print(embedded[1:2])
```

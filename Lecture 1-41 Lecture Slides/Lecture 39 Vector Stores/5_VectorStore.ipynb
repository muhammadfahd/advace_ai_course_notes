## From loading - Embedding

```python
!pip install langchain langchain_community
from langchain.text_splitter import RecursiveCharacterTextSplitter
!pip install transformers langchain_huggingface
```

```python
!pip install pypdf
from langchain.document_loaders import PyPDFLoader
```

```python
loaders=PyPDFLoader('/content/FYP Report PhysioFlex(25july).pdf')
doc=loaders.load()
```

```python
textsplitter=RecursiveCharacterTextSplitter(
    chunk_size=300,
    chunk_overlap=50
)
split=textsplitter.split_documents(doc)
```

```python
from langchain_huggingface import HuggingFaceEmbeddings
embed_model = HuggingFaceEmbeddings(model_name='BAAI/bge-small-en-v1.5')

```

```python
embeddedquery=embed_model.embed_query("what is the scope of this project")
print(embeddedquery)
```

```python
embedded=embed_model.embed_documents([doc.page_content for doc in split])
```

```python
print(embedded[1:2])
```

### setting up chroma as vector store

```python
!pip install chromadb
```

```python
from langchain.vectorstores import Chroma

# Initialize Chroma vector store
chroma_store = Chroma(embedding_function=embed_model, persist_directory="chroma_db")

# Add documents to Chroma
chroma_store.add_texts([doc.page_content for doc in split])

# Save the Chroma vector store
chroma_store.persist()
```

```python
!pip install langchain langchain_community qdrant-client transformers langchain_huggingface pypdf

from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.document_loaders import PyPDFLoader
from langchain_huggingface import HuggingFaceEmbeddings
from qdrant_client import QdrantClient
from langchain.vectorstores import Qdrant
from qdrant_client.http import models as rest


# Qdrant setup
url = "https://ad84b07d-d66f-49e1-9954-36216c0a88cf.europe-west3-0.gcp.cloud.qdrant.io/"
api_key = "QQB6t-CjTUL3GM2zZjQEXjzf5KhDS5ce3i_RU5zjRXTWP_Oni-oeQg"
qdrant_client = QdrantClient(url=url, api_key=api_key)
collection_name = "fypsample"

# Recreate collection in Qdrant
qdrant_client.recreate_collection(
    collection_name=collection_name,
    vectors_config=rest.VectorParams(size=len(embedded[0]), distance=rest.Distance.COSINE)
)

# Prepare and upload documents to Qdrant
documents = [{"id": i, "text": doc.page_content, "embedding": embedding}
             for i, (doc, embedding) in enumerate(zip(split, embedded))]

qdrant_client.upload_collection(
    collection_name=collection_name,
    vectors=[doc["embedding"] for doc in documents],
    payload=[{"page_content": doc["text"]} for doc in documents],
    ids=[doc["id"] for doc in documents],
    batch_size=50
)

# LangChain Qdrant vector store
vector_store = Qdrant(client=qdrant_client, collection_name=collection_name, embeddings=embed_model)

```

```python
# Your query as a string
query = "who is moniza"

# Generate the embedding for the query (pass the query as a string, not a list)
query_embedding = embed_model.embed_query(query)

# Perform similarity search using the generated query embedding
results = vector_store.similarity_search(query)
print(results)

```
